# Customer Retention and Exploratory Data Analysis

```python
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Load the data
customer = pd.read_csv(r'..\Data\churn_raw.csv')
```

## Summary

```python
# Data Types and Missing Values
print(customer.info(verbose=True))
```

```python
# Summary Statistics
print(customer.describe())
```

## Visualize Data

```python
# Distribution of numerical variables
customer.hist(bins=30, figsize=(20,15))
```

```python
# Frequency counts for categorical variables
for column in customer.select_dtypes(include=['object']).columns:
    print(customer[column].value_counts())
```
